```python
import cv2
import matplotlib.pyplot as plt

image=cv2.imread(r"C:\Users\Admin\Desktop\Datasets\Dl\CNN\jpg",cv2.IMREAD_GRAYSCALE)

#Apply Canny Edge_detection
edges=cv2.Canny(image,threshold1=0,threshold2=200)
plt.figure(figsize=(10,5))

#original Image
plt.subplot(1,2,1)
plt.imshow(image,cmap='gray')
plt.title('Original Gray Image')
plt.axis('off')

#Edge_detected Image
plt.subplot(1,2,2)
plt.imshow(edges,cmap='gray')
plt.title('Edges Detection')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image=cv2.imread(r"C:\Users\Admin\Desktop\Datasets\Dl\CNN\jpg")

image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
grey_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

edges=cv2.Canny(grey_image,threshold1=100,threshold2=200)

plt.figure(figsize=(10,5))

#Original Color Image
plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title('Original Color Image')
plt.axis('off')

#Edge_detected Image
plt.subplot(1,2,2)
plt.imshow(edges,cmap='gray')
plt.title('Edges Detection')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
 
# Step 1: Read the image
image = cv2.imread(r"C:\Users\Admin\Desktop\Datasets\Dl\CNN\jpg")
 
# Step 2: Split the image into its color channels (BGR format)
blue_channel, green_channel, red_channel = cv2.split(image)
 
 
 
# Create a blank image with the same size as the original
blank = np.zeros_like(blue_channel)

blue_only= cv2.merge([blue_channel,blank ,blank ]) #keep only blue channel
green_only= cv2.merge([blank,green_channel ,blank ])#keep only green channel
red_only= cv2.merge([blank ,blank ,green_channel])#keep only red channel

#Detect Image
plt.figure(figsize=(10,7))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

#for blue clr
plt.subplot(1, 4, 2)
plt.imshow(blue_only)
plt.title("Blue Channel")
plt.axis("off")

#For green color
plt.subplot(1, 4, 3)
plt.imshow(green_only)
plt.title("Green Channel")
plt.axis("off")

# for red color
plt.subplot(1, 4, 4)
plt.imshow(red_only)
plt.title("Red Channel")
plt.axis("off")

plt.tight_layout()
plt.show()
```
